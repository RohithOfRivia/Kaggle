```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("data/CarTrade.csv")
df.head(10)
```

```python
df.dtypes
```

```python
#Checking for null values
issue_df = df[df.isnull().any(axis=1)]
issue_df.head()
```

```python
#Removing the comma from some columns to convert them into int
df['Car_Price'] = df['Car_Price'].apply(lambda x: x.replace(',', ''))
df['Car_Price'] = df['Car_Price'].astype('int64')
df.head()
```

```python
df['Car_Kms'] = df['Car_Kms'].apply(lambda x: x.split(' ')[0].replace(',', ''))
df.head(10)
```

```python
df['Car_Kms'] = df['Car_Kms'].astype('int64')
```

```python
df.dtypes
```

```python
df.describe()
```

```python
cars = df['Car_Model'].unique()
cars[0].strip('[]').strip('\'\' ')
```

```python
df['Car_Model'] = df['Car_Model'].apply(lambda x: x.strip('[]').strip('\'\' '))
df.head(10)
```

```python
df['Car_Company'] = df['Car_Model'].apply(lambda x: x.split(' ')[0])

df['Car_Company'] = df['Car_Company'].replace({'Land': 'Land Rover', 'MG': 'MG Hector'})
```

```python
df.head()
```

```python
df['TrueLoc'] = df['Car_Location'].apply(lambda x: x.split(' ')[0].strip())

df.head()
```

```python
df['TrueLoc'] = df['TrueLoc'].apply(lambda x: x.replace('Navi', 'Navi Mumbai'))
```

```python
df['TrueLoc'].unique()
```

```python
df['model'] = df['Car_Model'].apply(lambda x: x.split(' '))
df['model'] = df['model'].apply(lambda x: ' '.join(x[1:2]))
df.head()
```

```python
df.loc[df['Car_Company'] == 'Land Rover', 'model'] = df['Car_Model'].apply(lambda x: x.split(' '))
df.loc[df['Car_Company'] == 'Land Rover', 'model'] = df['model'].apply(lambda x: ' '.join(x[2:5]))
df.head()
```

```python
from category_encoders import TargetEncoder
final_df = df.copy()
```

```python
tencoder = TargetEncoder()
final_df['Car_Model'] = tencoder.fit_transform(final_df['Car_Model'], final_df['Car_Price'])
final_df.head()
```

```python
def thedummies(col, dff):
    temp = pd.get_dummies(dff[col], drop_first=True)
    dff = dff.drop(col, axis=1)
    dff = pd.concat([dff, temp], axis=1)
    return dff
```

```python
final_df = thedummies('Car_Fuel', final_df)
final_df = thedummies('Car_Company', final_df)
final_df.head()
```

```python
final_df = final_df.drop(['Car_Location', 'TrueLoc', 'model'], axis=1)
final_df.head()
```

```python
from sklearn.linear_model import LinearRegression
import sklearn.preprocessing
from sklearn.model_selection import train_test_split
```

```python
X = final_df.drop('Car_Price', axis=1).values
X = sklearn.preprocessing.scale(X)
y = final_df['Car_Price'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.7)
```

```python
clf = LinearRegression()
clf.fit(X_train, y_train)
```

```python
clf.score(X_test, y_test)
```

```python
for xtest, ytest in zip(X_test, y_test):
    print(f'Actual price: {ytest} Forecast price: {clf.predict([xtest])[0]}')
```
